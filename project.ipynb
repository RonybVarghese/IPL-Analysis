```python
df1['extras_type'] = df1['extras_type'].fillna('None')
df1['player_dismissed'] = df1['player_dismissed'].fillna('None')
df1['dismissal_kind'] = df1['dismissal_kind'].fillna('None')
df1['fielder'] = df1['fielder'].fillna('None')
```

# Cleaning the df2 dataset

```python
df2['city'] = df2['city'].fillna('Unknown')
df2['result_margin'] = df2['result_margin'].fillna(0)
df2['target_runs'] = df2['target_runs'].fillna(0)
df2['target_overs'] =df2['target_overs'].fillna(0)
df2['method'] = df2['method'].fillna('Normal')
```

# Renaming 'id' to 'match_id

```python
df2.rename(columns={'id': 'match_id'}, inplace=True)
```

# Merging the datasets on 'match_id'

```python
merged_data = pd.merge(df1, df2, how='inner', on='match_id')
```

# 1.Calculate the number of matches each team has played

```python
team_matches_played = merged_data.groupby('team1')['match_id'].nunique() 
+ merged_data.groupby('team2')['match_id'].nunique()
team_matches_played
```

##### Conclusion:
##### The Number of matches played team is Royal Challengers Banglore(RCB)

# 2.Calculate the number of matches each team has won

```python
team_wins = df2.groupby('winner')['match_id'].nunique()
team_wins
```

# 3.Calculate the winning percentage

```python
team_win_percentage = (team_wins / team_matches_played) * 100
team_win_percentage.round(2)
```

# 4.Sort by winning percentage

```python
team_win_percentage = team_win_percentage.sort_values(ascending=False)
team_win_percentage.round(2)
```

##### Conclusion:
##### Gujarat Titans is higher win percentage in the ipl 

# 5.Calculate total runs scored by each batsman

```python
batsman_runs = merged_data.groupby('batter')['batsman_runs'].sum()
batsman_runs
```

# 6. Get the top 10 batsmen

```python
top_batsmen = batsman_runs.sort_values(ascending=False).head(10)
top_batsmen
```

# 7. Plot the top 10 batsmen


```python
plt.figure(figsize=(12, 8))
top_batsmen.plot(kind='bar', color='skyblue')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Batsmen')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Conculsion:
##### The captain of Royal Challengers Bangalore (RCB), Kohli holds the record for the highest number of runs in IPL history.

# 8. Filter deliveries where a wicket has fallen


```python
wickets = merged_data[merged_data['is_wicket'] == 1]
wickets
```

# 9. Calculate total wickets taken by each bowler

```python
bowler_wickets = wickets.groupby('bowler')['is_wicket'].sum()
bowler_wickets
```

# 10. Get the top 10 bowlers

```python
top_bowlers = bowler_wickets.sort_values(ascending=False).head(10)
top_bowlers
```

# 11. Plot the top 10 bowlers

```python
plt.figure(figsize=(12, 8))
top_bowlers.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Bowlers by Total Wickets')
plt.xlabel('Bowlers')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Conculsion:
##### Top 10 Bowler list DJ Bravo has been a key wicket-taker

# 16. Match venue ground

```python
matches=df2[df2['target_runs']>150]
matches=matches['venue'].value_counts().head(20).reset_index()
matches.head()
```

# 17. Runs scored>150

```python
matches['count'] = matches.groupby('venue')['venue'].transform('count')
plt.figure(figsize=(13,5))
sns.barplot(x=matches['venue'], y=matches['count'], palette="dark:b")
plt.xticks(rotation='vertical', size=7)
plt.title("Ground Statistics\n", size=12)
plt.suptitle("\n\n(runs scored>150)", size=10)
plt.ylabel("Count")
plt.xlabel("Venue")
plt.show()
```

##### Conclusion:
##### 150> scored top ground is Wankhede Stadium
